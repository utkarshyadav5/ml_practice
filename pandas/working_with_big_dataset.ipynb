##### To visualize multiple variables, boxplots are useful, especially when one of the variables is categorical.

```python
job['Initial Cost']

# we can see that initial cost is a string thus we cant plot string with string one data has to be numeric
```

```python
# first we remove $ than convert it into float
job['Initial Cost']=job['Initial Cost'].str.replace('$','')
job['Initial Cost']=pd.to_numeric(job['Initial Cost'])
job['Initial Cost']
```

```python
# boxplot where y is a numeric value and x is non numeric

job.boxplot(column='Initial Cost',by='Borough',rot=90)
plt.show()
```

```python
job['Total Est. Fee']=job['Total Est. Fee'].str.replace('$','')
job['Total Est. Fee']=pd.to_numeric(job['Total Est. Fee'])
job['Total Est. Fee']
```

```python
# When you want to visualize two numeric columns, scatter plots are ideal.

job.plot(kind='scatter', x='Initial Cost',y='Total Est. Fee',rot=70)
plt.show()
```
